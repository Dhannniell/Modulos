```python
# 1. Contador de palabras clave en los comentarios, excluyendo palabras cortas 

palabras = []
for r in resennas:
    palabras.extend([p for p in r["comentario"].lower().split() if len(p) > 3])

conteo_palabras = collections.Counter(palabras)
print("Palabras mas frecuentes:")
print(conteo_palabras.most_common(5))
```

```python
# 2. Agrupacion de reseñas por puntaje

grupos_resennas = collections.defaultdict(list)
for r in resennas:
    grupos_resennas[r["puntaje"]].append(r["comentario"])

print("Reseñas agrupadas por puntaje")    
for puntaje, comentarios in grupos_resennas.items():
    print(f"Puntaje:{puntaje}:{comentarios}")
```

```python
# 3. Generado por palabras clave
print("Combinacion de palabras claves con comentario")
combinaciones = itertools.combinations(conteo_palabras.keys(),2)
print(list(combinaciones))
```

```python
# 4. Generar una respuesta automatica

respuestas = ["Gacias por tu compra", "Esperamos verte pronto","Nos alegra que te haya gustado", "Hasta pronto","Apreciamos tu comentario"]
permutaciones_respuestas = itertools.permutations(respuestas, 2)
print("Respuestas automaticas")
for respuesta in permutaciones_respuestas:
    print("-".join(respuesta))
```

```python
plt.Figure(figsize=(8,5))
plt.bar(*zip(*conteo_palabras.most_common(5)))
plt.xlabel("palabras")
plt.ylabel("Frecuencia")
plt.title("Top 5 Palabras Mas Frecuentes En Reseñas")
plt.show()
```

```python
# 6. Grafica de pastel de puntajes.
puntajes = [r["puntaje"] for r in resennas]
conteo_puntajes = collections.Counter(puntajes)


plt.Figure(figsize=(12,12))

plt.pie(conteo_puntajes.values(), labels=conteo_puntajes.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Distribucion De Puntajes Por Reseñas")
plt.show()

```
